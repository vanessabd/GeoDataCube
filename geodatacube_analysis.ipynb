```python
datacube.dataset.rio.write_crs("EPSG:4326", inplace=True)
print(datacube.dataset.rio.crs)
```

```python
datacube.visualize_layer(variable="T2M", time_index=0, cmap="viridis")
```

```python
datacube.generate_heatmap(variable="T2M", time_index=0, cmap="coolwarm")
```

```python
mean_data = datacube.visualize_mean(variable="T10M")
print(mean_data)
```

```python
trend= datacube.calculate_trend(variable="T2M")
```

```python
metadata = datacube.get_metadata()
print(metadata)
```

```python
monthly_mean = datacube.aggregate_time(variable="T2M", method="mean", freq="ME")
print(monthly_mean)

```

```python
monthly_min = datacube.aggregate_time(variable="T2M", method="min", freq="ME")
print(monthly_min)
```

```python
resampled_data = datacube.resample_resolution(variable ="T2M", factor=10)
print(resampled_data)
```

```python
anomalies = datacube.compute_anomalies(variable="T2M", baseline ="none")
print(anomalies)
```

```python
resampled_data = datacube.resample_resolution(variable ="T2M", factor=5)
print(resampled_data)
```

```python
##comparison of the values compared to a baseline (if the baseline is none it considerates the mean over time)
anomalies = datacube.compute_anomalies(variable="T2M", baseline = 20)
print(anomalies)
```

```python
masked_data = datacube.apply_mask( variable="T2M", min_value=260, max_value=310)
print(masked_data)
```
